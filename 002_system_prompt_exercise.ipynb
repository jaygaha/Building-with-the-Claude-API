```python
# Load environment variables
from dotenv import load_dotenv

load_dotenv()
```

```python
# Create an API client
from anthropic import Anthropic

client = Anthropic()

model = "claude-sonnet-4-0"
```

```python
# Helpers
def add_message(messages, content, role):
    message = {"role": role, "content": content}
    messages.append(message)

def chat(messages, system=None):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
    }

    if system:
        params["system"] = system

    message = client.messages.create(**params)

    return message.content[0].text
```

```python
messages = []

"""
QUESTION:
-> Make claude to respond as concise as possible.
"""

system_prompt = """
# You are a expert Python programmer.
# Answer the user's questions with clear, short and concise code examples.
You are a Python engineer who writes very concise code.
"""

add_message(messages, 
            "Write a Python function that checks a string for duplicate characters.", 
            role="user")
answer = chat(messages, system=system_prompt)

answer
```
