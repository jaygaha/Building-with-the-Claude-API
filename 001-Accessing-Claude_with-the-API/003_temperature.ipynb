```python
"""
TEMPERATURE
-> It controls the randomness of the model's output.
-> A low temperature (e.g., 0.2) makes the output more focused and deterministic.
-> A high temperature (e.g., 0.8) makes the output more diverse and creative.
"""

# Load environment variables
from dotenv import load_dotenv

load_dotenv()
```

```python
# Create an API client
from anthropic import Anthropic

client = Anthropic()

model = "claude-sonnet-4-0"
```

```python
# Helpers
def add_message(messages, content, role):
    message = {"role": role, "content": content}
    messages.append(message)

def chat(messages, system=None, temperature=1.0):
    params = {
        "model": model,
        "max_tokens": 1000,
        "messages": messages,
        "temperature": temperature,
    }

    if system:
        params["system"] = system

    message = client.messages.create(**params)

    return message.content[0].text
```

```python
messages = []

add_message(messages, 
            "Generate a one sentence movie idea", 
            role="user")
answer = chat(messages, temperature=1.0)

answer
```
